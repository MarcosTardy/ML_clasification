1. **Precisión (Precision)**: La precisión mide la proporción de instancias positivas que el modelo clasificó correctamente como positivas en relación con todas las instancias clasificadas como positivas. Es una métrica importante cuando queremos minimizar los falsos positivos.

2. **Recall (Recall o Sensibilidad)**: El recall mide la proporción de instancias positivas que el modelo clasificó correctamente como positivas en relación con todas las instancias reales que son positivas. Es una métrica importante cuando queremos minimizar los falsos negativos.

3. **F1-Score**: El F1-Score es la media armónica de precisión y recall. Es útil cuando se busca un equilibrio entre la precisión y el recall. Es especialmente útil cuando las clases no están equilibradas en el conjunto de datos.

4. **Soporte (Support)**: El soporte es el número de muestras en cada clase real. Es útil para comprender cuántas muestras pertenecen a cada clase en el conjunto de prueba.

5. **Exactitud (Accuracy)**: La exactitud mide la proporción de todas las predicciones correctas (tanto positivas como negativas) en relación con todas las predicciones realizadas por el modelo. Es una métrica útil cuando las clases están equilibradas, pero puede ser engañosa en conjuntos de datos desequilibrados.

6. **Macro Promedio (Macro Avg)**: Esta fila proporciona el promedio no ponderado de las métricas de precisión, recall y F1-Score en todas las clases. Trata cada clase por igual, independientemente de su tamaño.

7. **Promedio Ponderado (Weighted Avg)**: Esta fila proporciona el promedio ponderado de las métricas de precisión, recall y F1-Score en todas las clases. Le da más peso a las clases con más instancias.

8. **Promedio de Soporte (Avg / Total)**: Esta fila muestra el promedio del soporte de todas las clases.

La segona i la tercera classe son les que pitjors mètriques obtenen, això es degut a que als dos models els costa distingir entre les classes ja que tenen característiques semblants. En canvi, la primera classe no mostra problemes.

#### 3.1. Model KNN

```python
classifier.get_params()
```

```python
from sklearn.metrics import f1_score

f1s = []

# Calculating f1 score for K values between 1 and 40
for i in range(1, 40):
    knn = KNeighborsClassifier(n_neighbors=i)
    knn.fit(X_train, y_train)
    pred_i = knn.predict(X_test)
    # using average='weighted' to calculate a weighted average for the 4 classes 
    f1s.append(f1_score(y_test, pred_i, average='weighted'))
    
plt.figure(figsize=(12, 6))
plt.plot(range(1, 40), f1s, color='red', linestyle='dashed', marker='o',
         markerfacecolor='blue', markersize=10)
plt.title('F1 Score K Value')
plt.xlabel('K Value')
plt.ylabel('F1 Score')

#kneelocator
```

```python
classifier2 = KNeighborsClassifier(n_neighbors=15)
classifier2.fit(X_train, y_train)
y_pred2 = classifier2.predict(X_test)
print(classification_report(y_test, y_pred2))
```

Tal i com es pot veure al gràfic de més adalt, el valor de k més óptim en el conjunt de dades d'entrenament, per tant que té una precisió major, és el 30 cap endavant. Significa que el model es fixa en els 30 valors més propers per fer la seva predicció. Malgrat això, no vol dir que sigui el resultat més óptim a l'hora de tractar amb dades que no s'han vist abans, es a dir, en el conjunt de dades de proba o en dades de la vida real. Utilitzaré GridSearch per comprovar-ho. Aquest model té com a objectiu trobar el conjunt d'hiperparàmetres que generalitzen millor en dades que no s'han vist abans.

```python
knn = KNeighborsClassifier() 
#Torno a fer el Random Forest ja que no vull fer grid search amb el model ja entrenat.
```

## Exercici 5
### Aplica algun procés d'enginyeria per millorar els resultats (normalització, estandardització, mostreig...)

#### 5.1. Scaling

```python
from scipy.stats import shapiro #Llibreria de python importa normaltest per fer l'arlgorisme de Shapiro.

columnes=X.columns.tolist() #emmagatzema una llista amb els noms de les columnes del X_train
for i in columnes: #Fa tants tests d'hip com columnes hi hagin.
    stat, p = shapiro(X[i]) 
    #L'algorisme shapiro assigna automàticament l'estadistica del test i el p-value.
    print('stat=%.3f, p=%.3f' % (stat, p))
    if p > 0.05: # L'alpha o nivell de significància és 0,05. Estem al 95% segurs...
     print('La variable', i,' probablement segueix una distribució Normal') 
    #Si el p-value és major que alpha, no rebutjem la hipòtesi nul·la
    else:
     print('La variable ', i, ' probablement NO segueix una distribució Normal') 
    # Si és més petit, rebutjem la hipòtesi nul·la.
```

Sembla que totes les variables seguiexen una distribusió no gaussiana excepte la variable "Alcalinity of ash", que se li aplicarà una estandarització.
